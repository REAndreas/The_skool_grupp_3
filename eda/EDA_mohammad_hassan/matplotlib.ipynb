```python
import duckdb

# Query example: Get top 5 fields by total students
duckdb.query("""
SELECT "Utbildningsområde", 
       SUM("Platser med start 2020" + "Platser med start 2021") AS total_students
FROM df
GROUP BY "Utbildningsområde"
ORDER BY total_students DESC
;
""").df()
```
