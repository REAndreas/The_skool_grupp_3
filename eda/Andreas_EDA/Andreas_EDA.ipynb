# EDA Yrkeshögskolan

```python
import pandas as pd

df = pd.read_csv("data/stud_utbildningsområde_övertid.csv")

df.head()
```

```python
df.info()
```

```python
df = df.drop(labels="kön", axis=1)

df
```

```python
df= df.drop(labels="ålder", axis=1)

df
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Transpose so years are on x-axis
df_plot = df.set_index("utbildningens inriktning").T

# Plot
df_plot.plot(kind="bar", figsize=(12, 6))


plt.title("Antal studerande per utbildningsinriktning (2020–2024)")
plt.xlabel("År")
plt.ylabel("Antal studerande")
plt.xticks(rotation=0)
plt.legend(title="Inriktning", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
df_plot.to_csv("data/Yrkeshogskolan_transformed.csv")
```
