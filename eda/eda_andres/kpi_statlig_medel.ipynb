```python
df_total = df[df['Utbildningsområde'] != 'Totalt']
df_total
```

```python
def total2024(df):
    df_total = df[df['Utbildningsområde'] != 'Totalt']
    return df_total['2024'].sum()

total_results_2024 = total2024(df)
print(total_results_2024)
    
```

```python
def total2023(df):
    df_total = df[df['Utbildningsområde'] != 'Totalt']
    return df_total['2023'].sum()

total_results_2023 = total2023(df)
print(total_results_2023)
```

```python
def total2022(df):
    df_total = df[df['Utbildningsområde'] != 'Totalt']
    return df_total['2022'].sum()

total_results_2022 = total2022(df)
print(total_results_2022)
```

```python
def total2021(df):
    df_total = df[df['Utbildningsområde'] != 'Totalt']
    return df_total['2021'].sum()

total_results_2021 = total2021(df)
print(total_results_2021)
```

```python
def total2020(df):
    df_total = df[df['Utbildningsområde'] != 'Totalt']
    return df_total['2020'].sum()

total_results_2020 = total2020(df)
print(total_results_2020)
```

```python
df_total_years = total_results_2020 + total_results_2021 + total_results_2022 + total_results_2023 + total_results_2024
print(df_total_years)
```
